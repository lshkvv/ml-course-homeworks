### We drop sepal width cause it is useless and we should fit into 3 dimensions

```python
viz_pd_Iris = pd_Iris.drop(columns='sepal width (cm)')
viz_pd_Iris.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D  
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")

x1 = viz_pd_Iris.iloc[:, 0]
x2 = viz_pd_Iris.iloc[:, 1]
x3 = viz_pd_Iris.iloc[:, 2]
y  = viz_pd_Iris["target"]

colors = {0: "red", 1: "green", 2: "blue"}

for cls, color in colors.items():
    mask = (y == cls)
    ax.scatter(
        x1[mask],
        x2[mask],
        x3[mask],
        c=color,
        label=f"{iris.target_names[cls]}",
        alpha=0.7,
    )

ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[2])
ax.set_zlabel(iris.feature_names[3])
ax.set_title("flowers")
ax.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

### We will classify only setosa/not setosa it is linear separeted

```python
y_iris = y_iris.apply(lambda z: z==0)
```

```python
X_iris = pd_Iris.drop(columns=['sepal width (cm)','target'])
```

```python
X_iris_train, X_iris_test,y_iris_train, y_iris_test = train_test_split(X_iris, y_iris,test_size=0.3,shuffle=True)
```

```python
model = LogisticRegression()
```

### Scaling isnt necessary 

```python
model.fit(X_iris_train,y_iris_train) 
```

```python
y_preds = model.predict(X_iris_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_iris_test, y_preds)
```
