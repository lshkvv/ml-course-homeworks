```python
import sys
import numpy as np
import pandas as pd
import sklearn

print("Python:", sys.version)
print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
print("sklearn:", sklearn.__version__)
```

```python
np.random.seed(42)
X = np.random.randn(100, 5)

# статистики по всей матрице
print("mean:", X.mean(), "std:", X.std(), "min:", X.min(), "max:", X.max())

# статистики по каждому столбцу
print("col mean:", X.mean(axis=0))
print("col std:", X.std(axis=0))

# векторизация
w = np.random.randn(5)
y = X @ w
print("y shape:", y.shape)
```

```python
cols = [f"feature_{i}" for i in range(5)]
df = pd.DataFrame(X, columns=cols)
df["target"] = y

print(df.head())
print(df.shape)
print(df.describe())
print(df.isna().sum())

# пара операций
df_filtered = df[df["feature_0"] > 0]
df_sorted = df.sort_values("target", ascending=False)
```
