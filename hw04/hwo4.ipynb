```python
corr = cats.corr(numeric_only = True)
```

```python
plt.figure(figsize = (8, 6))
sns.heatmap(
    corr,
    annot = True,
    cmap="PuRd",
    vmin=-1, vmax=1
)
plt.title("Cats correlation heatmap ")
plt.show()
```
