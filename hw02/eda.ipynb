```python
data = data[data["Age_in_months"] > 0]
```

```python
data["Age_in_years"] = data["Age_in_years"].fillna(data["Age_in_months"]/12)
```

```python
data = data.dropna(subset = ["Gender", "Breed", "Neutered_or_spayed", "Allowed_outdoor"])
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data["Preferred_food"] = data["Preferred_food"].fillna(data["Preferred_food"].mode()[0])
```

```python
data["Body_length"] = data["Body_length"].fillna(
    data.groupby(["Breed", "Age_in_years"])["Body_length"]
        .transform("median")
)
```

```python
data["Weight"] = data["Weight"].fillna(
    data.groupby(["Breed", "Age_in_years"])["Weight"]
        .transform("median")
)
```

```python
data = data.dropna(subset = ["Body_length", "Weight"])
```

```python
data["Fur_pattern"] = data["Preferred_food"].fillna(data["Preferred_food"].mode()[0])
```

```python
mode_by_breed = data.groupby("Breed")["Fur_pattern"].transform(
    lambda s: s.mode()[0]
)

data["Fur_pattern"] = data["Fur_pattern"].fillna(mode_by_breed)
```

```python
mode_by_breed = data.groupby("Breed")["Fur_pattern"].transform(
    lambda s: s.mode()[0]
)

data["Fur_pattern"] = data["Fur_pattern"].fillna(mode_by_breed)
```

```python
data["Owner_play_time_minutes"] = data["Owner_play_time_minutes"].fillna(
    data.groupby("Breed")["Owner_play_time_minutes"].transform("median")
)
```

```python
data["Eye_colour"] = data["Eye_colour"].fillna(
    data.groupby("Breed")["Eye_colour"].transform(lambda s: s.mode()[0])
)
```

```python
data["Sleep_time_hours"] = data["Sleep_time_hours"].fillna(
    data.groupby("Breed")["Sleep_time_hours"].transform(lambda s: s.mode()[0])
)
```

```python
data["Country"] = data["Country"].fillna(
    data.groupby(["Latitude", "Longitude"])["Country"].transform("first")
)
```

```python
data["Latitude"] = data["Latitude"].fillna(
    data.groupby("Country")["Latitude"].transform("median")
)
```

```python
data["Longitude"] = data["Longitude"].fillna(
    data.groupby("Country")["Longitude"].transform("median")
)
```

```python
data = data.dropna()
```
